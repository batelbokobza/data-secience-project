# <font color=royalblue>Algorithm Description</font>

The algorithm for identifying the speaker is based on the following steps - receiving a voice recording, 
analyzing existing voice features, and turning a  Fourier Transform into a suitable format that should get a good match
percentage between the received voice, the existing voices in the program templates, 
and finally get the match result.<br>

The speaker recognition algorithm consists of two steps: registration and verification.<br>
At the registration stage, recorded samples of the speakers voice are made, 
and analyzed to determine his voice imprint. In the verification phase, 
the speech sample is compared against a number of sound impressions to determine the best match.<br>

The sound recognition algorithm uses a discrete Fourier transform as stated, 
to compare the frequency spectrum of two sound samples.<br>
The conversion for comparison of the two recordings is done by finding the Euclidean norm.<br>
Convert the recording to a frequency range, which can be captured in a 600-sized vector in the Euclidean space.<br>
Once the norm is found, a comparison can be made between the two vectors of the voices by 
normalizing them, and calculating the norm of the difference between them.<br>
In order to determine whether two votes belong to the same person, inequality is used in Chebyshev.<br>
Inequality basically tells us that for any positive number K, at least 1- (1-(k ^ 2)) data will be 
within k deviations from the mean.<br>

The purpose of a Fourier transform is basically to convert an audio unit represented by a sound wave from the time domain, to the frequency domain.<br>
This process is called a spectrogram, and we will expand on it later.<br>

In conclusion, the voice recognition algorithm requires the following design to complete voice 
recognition with maximum success:<br>

***Step 1*** - Create a group of sample voice recordings of different voices, for comparison with the voice that needs to be identified. <br>
The number of samples of each sound should be sufficient to make the comparison. <br>
we will create our data set by creating a set of sample voice recordings of different speakers.<br>

***Step 2***- Cleaning, preparation and arrangement of the data.<br>
Create folders to be used for training, testing and verifying the algorithm.<br>

***Step 3*** - Selecting and defining the features and characteristics that our model will use to evaluate audio files, and find and extract the features for each audio file and build a feature matrix that represents all the features of all the audio files.<br>


***Step 4*** - building the model and evaluating it for the purpose of assessing compatibility with the data.
Extract a matching result from the algorithm for two speech segments. <br>
Comparing two speech segments represented by the identity vectors created in step 3, 
and calculating the results of the comparison between the unfamiliar voice segment obtained 
and the other voice segment recognized in the data set.<br>

***Step 5*** - Matching templates for comparison that the same algorithm will need to perform. <br>
Decide whether the two audio files belong to the same speaker or not. <br>
It's performed by a predetermined threshold value - if the result is greater than it, the answer is yes. <br>
Otherwise, it is negative.<br>

```python
def get_directory_data_paths(path):
    return tf.io.gfile.glob(path)
```

```python
training_speakers_paths = get_directory_data_paths('datasets/LibriSpeech/train-clean-100/*')
```

We want to create three sets of data for training, validation and testing of the algorithm.<br>
In the next function, for each speaker we will add 4 sound segments belonging to him to the training set, and to the verification set and test set we will add 3 sound segments belonging to him.<br>
We chose to select the sound segments that are 15 seconds long.

```python
def arrange_and_get_speaker_clip_sets(speaker_clips):
    speaker_clips = get_directory_data_paths(speaker_path + '/*/*')
    training_clips = []
    validation_clips = []
    test_clips = []
    for clip in speaker_clips:
        if '.txt' not in clip:
            if 14 < librosa.get_duration(filename=clip) <= 15:
                if len(training_clips) != 4:
                    training_clips.append(clip)
            elif len(test_clips) != 3:
                test_clips.append(clip)
            elif len(validation_clips) != 3:
                validation_clips.append(clip)
            else:
                break
    return {'training_clips': training_clips, 'test_clips': test_clips, 'validation_clips': validation_clips}
```

We will go over the data in the data set, and save 30 different speakers, with their sound clips.

```python
speakers_sets = dict()
for speaker_path in training_speakers_paths:
    speaker_set = arrange_and_get_speaker_clip_sets(speaker_path)
    if len(speaker_set['training_clips']) == 4 and len(speaker_set['test_clips']) == 3 and len(speaker_set['validation_clips']) == 3:
        speaker_name = speaker_path.split('datasets\\LibriSpeech\\train-clean-100\\')[1]
        speakers_sets[speaker_name] = speaker_set
        if len(speakers_sets) == 30:
            break
```

# <font color=lightskyblue>1. Spectrogram</font>

A spectrogram is a visual way of representing the signal strength, or "volume", of a signal over time at different frequencies that are in a particular waveform.<br>
Is a visual representation of the frequency spectrum of a signal as it changes over time.<br>
or in other words, It is a process of visual representation of all the core components of audio data (time, frequency and size).<br>

A spectrogram explains how the signal strength is distributed in every frequency found in the signal.<br>
It transmits the signal strength through the colors - the lighter the color the higher the energy of the signal
One axis represents time, and the other axis represents frequency. <br>
A third dimension indicating the amplitude of a particular frequency at a particular time is represented by the 
intensity or color of each point in the image.<br>

When we talk about sound, we usually talk about a sequence of oscillations at different pressure intensities, so the saying - to imagine a sound means - to imagine air waves.<br><br>

The sound wave we saw gives us only a two-dimensional representation of the sound segments of the speakers.<br>
Sound can be represented in another mathematical representation called Fourier transformation.<br>
Fertile transformation is a function that receives a signal in the time domain as input, and produces its decomposition into frequencies.<br>
He decomposes a function of time (signal) into constituent frequencies. <br>
In the same way a musical chord can be expressed by the volumes and frequencies of its constituent notes, a Fourier Transform of a function displays the amplitude (amount) of each frequency present in the underlying function (signal).
<br>
This is what a digital signal frequency wave looks like:<br>

Let's see what we get from applying the Fourier transform on the speaker's sound wave we saw earlier:<br>

```python
import matplotlib.pyplot as plt

n_fft = 2048
D = np.abs(librosa.stft(x_one[:n_fft], n_fft=n_fft, hop_length=n_fft+1))
plt.title("\nFourier transform on the sound wave:\n", color='blue', fontsize=15, fontweight="bold")
plt.plot(D);
```

Let's see what we get when we try to apply the Fourier transform on a time window to transfer a waveform from a time domain to a frequency domain:

As we said before, the STFT feature converts data to short-term fertility conversion.<br>
The STFT basically converts signals of the sound waves so that we can know what the amplitude of the given frequency is at a given time.<br> 
Using the STFT feature we can determine the amplitude of the different frequencies played at a given time of an audio signal.
<br>Specshow is used to display a spectrogram.<br>

```python
import librosa.display
hop_length = 512
D = np.abs(librosa.stft(x_one, n_fft=n_fft,  hop_length=hop_length))
librosa.display.specshow(D, sr=Fs_one, x_axis='time', y_axis='linear');
plt.colorbar();
```

We seem to have got a black window, from which nothing can be seen or understood.<br>
This is because the sound that humans usually hear is concentrated in very small amplitude ranges.<br><br>
To get the spectrogram, we will make a small change in the window we received: <br>
We will turn the x-axis representing the frequency into a diary and the y-axis representing the amplitude will become decibels. <br>

```python
DB = librosa.amplitude_to_db(D, ref=np.max)
librosa.display.specshow(DB, sr=Fs_one, hop_length=hop_length, x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
```

Let's see what the representation of the short-range power spectrum of the sound waves we saw to the sounds of the speakers and the differences between them look like:

```python
X_one = librosa.stft(x_one)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(np.abs(X_one), ref=np.max), sr=Fs_one, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title("\nSpeaker #" + one_speaker_name + "\n", color='coral', fontsize=15, fontweight="bold")
plt.tight_layout()
```

# <font color=lightskyblue>2. MFCC - Mel-frequency cepstral coefficients</font>

The STFT is built using the MFCC feature.<br>
Humans' hearing is natural, and our ears act like a kind of filter.<br>
The MFCC is based on the bandwidth of the human ear according to the frequency of hearing.<br>
Between the ear filters there are linearly arranged intervals with low frequency, and logarithmically with high frequency.<br>
The purpose of these intervals is to grasp the properties of speech, and this is basically done by mel-scale.
The MFCC is a mathematical method that converts the power spectrum of an audio signal into a small number of coefficients that represent the volume of the audio signal from the time zone to the frequency zone.<br>
It basically produces the short-range power spectrum of the sound, based on a linear cosine transformation.<br>
When the MFCC feature achieves the Capstrom Mel frequency, it will allow better representation of the sound in audio compression.<br>

Let's see what a representation of the short - range power spectrum of this sound looks like on this feature:<br>

```python
from matplotlib.colors import Normalize

mfccs_one = librosa.feature.mfcc(x_one, sr=Fs_one)
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs_one, sr=Fs_one, x_axis='time', norm=Normalize(vmin=-30, vmax=30))
plt.colorbar()
plt.yticks(())
plt.ylabel('MFC Coefficient', fontsize=13)
plt.title("\nSpeaker #" + one_speaker_name + "\n", color='coral', fontsize=20, fontweight="bold")
plt.tight_layout()

mfccs_two = librosa.feature.mfcc(x_two, sr=Fs_two)
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs_two, sr=Fs_two, x_axis='time', norm=Normalize(vmin=-30, vmax=30))
plt.colorbar()
plt.yticks(())
plt.ylabel('MFC Coefficient', fontsize=13)
plt.title("\nSpeaker #" + two_speaker_name + "\n", color='coral', fontsize=20, fontweight="bold")
plt.tight_layout()
```

A positive MFFC value corresponds to regions with a low frequency of spectrum, and a negative value corresponds to a high frequency for regions with a hige frequency of a spectrum.<br>
It is basically a set of features that describe the general shape of the spectral envelope.<br>
The spectral envelope is the relative power of the various definitions that is influenced by different factors, and it is this that actually determines the “nature” of the sound, i.e. the characteristics of the human voice of the speakers.<br><br>

# <font color=lightskyblue>3. Mel Scale</font>

The Mel scale is the result of a non-linear transformation of the frequency scale.<br>
It is a perceptual scale of the sounds that are heard by the listeners as equals the distance from each other.<br>
The reference point between this scale and standard frequency measurement is defined by assigning a perception height of 1000 miles to a ton of 1000 Hz, 40 dB above the listener threshold.<br>
Above about 500 Hz, increasing intervals are examined by the listeners to produce equal intervals.<br>

To extract the audio properties we saw in the explanation above, we use a Python library called librosa.<br>
librosa is a python package for music and audio analysis, which enables audio analysis and feature extraction.<br>
It provides the building blocks necessary to create music information retrieval systems. <br>
<br>We will briefly mention the characteristics of the features: 
- The mfcc feature, will achieve the Capstrom Mel frequency, which allows better representation of the sound in audio compression. <br>It basically produces the short-range power spectrum of the sound, which is based on a linear cosine transformation.
- The chroma - stft feature, will calculate the waveform and spectrogram for the audio files. <br>This feature is based on a short-lived Fourier Transformation.<br>
- The feature melspectrogram, will calculate the spectrogram according to the audio units from our database. <br>
- The spectral - contrast feature is used to calculate spectral contrast We will achieve the decibel difference between peaks and valleys in the spectrum.<br>
- The trait tonnetz, calculates the tonal centroid properties, is a calculation of a weighted Fourier transposition with normalized chromium vectors.

<br>The following function will be used by us to extract the audio properties from all the clips that are in our data set:

```python
def extract_features(file_name):
    X, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0)  # array of length 40.
    stft = np.abs(librosa.stft(X))
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)  # array of length 12
    mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)  # array of length 128
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)  # array of length 7
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X), sr=sample_rate).T,axis=0)  # array of length 6
    return mfccs, chroma, mel, contrast, tonnetz
```

After extracting the features in each of the audio files, we copute using function we wrote feature matrix according to the extracted features:<br>
The function creates a matrix that consists of arrays that contain the features of the audio files.<br>

```python
def get_matrix_features(files_path, name_df):
    matrix_features = []
    for file_name in name_df['File']:
        file_path = files_path + file_name
        mfccs, chroma, mel, contrast, tonnetz = extract_features(file_path)
        matrix_features.append(np.concatenate((mfccs, chroma, mel, contrast, tonnetz), axis=0))
    return np.array(matrix_features)
```

At this point, when we have the desired x and y, we can build our feed forward neural network. <br>
A model of the neural network operating in feed forward was constructed. <br>

We will create the neural network model by Sequential to create a layeres model.<br>

**Note -** If you want to add or download data to the data set, or make any changes to the data set, you need to pay attention to their balance - the size of the network versus the size of the data.<br>
Our data set contains enough data for our network's learning process.<br>
If the data set is small and the network is large, the learning process can stop in the middle or give the same results multiple times without justification.<br>
And if the data set is large and the network is small, the loss function will jump significantly because the network capacity will not be able to store all the information.<br>

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.callbacks import EarlyStopping
model = Sequential()
model.add(Dense(193, input_shape=(193,), activation = 'relu'))
model.add(Dropout(0.1))
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.25))
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(30, activation = 'softmax'))
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
early_stop = EarlyStopping(monitor='val_loss', min_delta=0, patience=100, verbose=1, mode='auto')
```

The use of the 'relu' activation function is quite clear, since we want our revised linear activation function to return the maximum between 0 and the tensor. <br>
In addition, since we have 30 different speakers, and each audio file belongs to only one speaker out of all 30 different speakers, the output of the activation function will be 'Softmax' which converts value vectors to a probabilistic distribution between 0 and 1.<br>

For example, there may be people with very similar voices and this can lead to an error in the speaker's choice / decision to which the voice segment belongs, and in order for our neural network model to better decide who owns the audio file, the output of the function will be 'Softmax'.<br>

The softmax function was chosen to be used as the last layer of the neural network because the input values can be positive, negative, zero or greater numbers than one,
And the efficiency of the softmax function converts these values to values between 0 and 1, so that the sum of all the values in the given vector is equal to 1.<br>

The mathematical definition of the softmax function looks like this:

All zi values are the elements of the input vector and they can be any real value.<br>
The sigma that is at the bottom of the formula is actually normalization.<br>
This calculation ensures that the sum of all the output values of the function will be equal to 1, thus constituting a valid probability distribution.<br>

Selected Optimization Algorithm - Adam <br>
Is a version of stochastic gradient drop (SGD).<br>
Unlike SGD, which maintains a constant learning rate throughout each iteration of training, a person actually ‘adapts’, or changes the pace of learning by taking into account the moving averages of the first and second moment (mean and variance) of the slope in each iteration of training.<br><br>

We are now building the model and adapting our model to the training and verification data we have created:<br>
